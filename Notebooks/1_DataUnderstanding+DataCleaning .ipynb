# Data Cleaning

```python
#check missing values:
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
#Only required columns keep it is:
df = df[['Id', 'ProductId', 'Score', 'Time', 'Summary', 'Text']]
```

```python
#rename the columns :  Because;Clean professional column names
df.columns = ['review_id', 'product_id', 'rating', 'review_time', 'summary', 'review_text']
```

```python
#check null values 
df.isnull().sum()
```

Insights 
Some reviews may have missing summary . Missing values need to be handled before NLP processing.

```python
#handel missing values 
df['summary'] = df['summary'].fillna('No Summary')
```

```python
df.isnull().sum()
```

```python
#check duplicates:
df.duplicated().sum()
```

```python
#if duplicated is present then remove it 
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
#Unix time convert into date 
df['review_date'] = pd.to_datetime(df['review_time'], unit='s')
```

# Create Sentiment column  : 
4 or 5 rating → Positive
3 rating → Neutral
1 or 2 rating → Negative

```python
def get_sentiment(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 3:
        return "Neutral"
    else:
        return "Negative"

df['sentiment'] = df['rating'].apply(get_sentiment)
```

```python
#save the cleaned dataset:
df.to_csv("cleaned_Review_dataset.csv", index=False)
```

```python
The dataset was cleaned by selecting relevant columns, handling missing values, removing duplicates, converting timestamp into date format, and creating sentiment labels from rating scores.
```
