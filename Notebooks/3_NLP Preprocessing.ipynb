```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv("cleaned_Review_dataset.csv")
df.head()
```

```python
#Text cleaning function
def clean_text(text):
    text = str(text).lower()              #str(text) → text ko string banata hai, lower() → capital letters ko small karta hai
    text = re.sub(r'[^a-z\s]', '', text)  #re.sub(r'[^a-z\s]', '', text) → special characters/numbers remove karta hai
    text = re.sub(r'\s+', ' ', text)      #re.sub(r'\s+', ' ', text) → extra spaces remove karta hai
    text = text.strip()                  #strip() → starting/ending spaces remove karta hai
    return text                          #return text → cleaned text return karta hai
```

```python
#Create Clean review column  :   Har review par cleaning function apply hota hai.
df['clean_review'] = df['review_text'].apply(clean_text)
```

```python
#Remove Neutrals
df = df[df['sentiment'] != 'Neutral']
```

```python
#Take Sample data :  WHY?: Full dataset heavy hai. 20,000 rows project ke liye enough hain.
df = df.sample(20000, random_state=42)
```
